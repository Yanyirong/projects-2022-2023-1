```python
def cutoff(num):
    if num >= 0:
        pn_flag = 0
    else:
        pn_flag = 1
    flag = num & 0x80
    if pn_flag == 0:
        if flag > 0:
            num = (num >> 8) + 1
        else:
            num = num >> 8
        if num > 127:
            num = 127
    else:
        if flag > 0 and num & 0x7F > 0:
            num = (num >> 8) + 1
        else:
            num = num >> 8
        if num < -128:
            num = -128
    return num
```

```python
output_file = open('cutoff_data.txt','w')
cutoff_data = np.zeros(4096, dtype='int8')
cutoff_data = cutoff_data.reshape([64,64])
```

```python
for i in range (output_data.shape[0]):
    for j in range (output_data.shape[1]):
        cutoff_data[i][j] = cutoff(output_data[i][j])
        output_file.write(hex(int(np.binary_repr(cutoff_data[i][j], width=8), 2))[2:].zfill(2) + '\n')
```
